```python
# importing libraries
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\manoj.c\Documents\PANDAS_PROJECT\Pandas-Project-2\Datasets\lego_sets.csv")
df.head()
```

```python
theme = pd.read_csv(r"C:\Users\manoj.c\Documents\PANDAS_PROJECT\Pandas-Project-2\Datasets\parent_themes.csv")
theme.head(10)
```

```python
# merging data using 1st table columns with 2nd table column names.
merged = df.merge(theme, left_on='parent_theme', right_on='name')
merged = merged.drop(columns= 'name_y')
merged.head()
```

```python

# Taking data only where is_licensed is True
licensed = merged[merged["is_licensed"]]
licensed.head()
```

```python
# gettinng data of only Star_wars
star_wars = licensed[licensed['parent_theme'] == "Star Wars"]
star_wars.head(10)
```

```python
# Shapes will get me number of rows & columns
star_wars.shape
```

Question 01: What percentage of all licensed sets ever released were Star Wars themed?
Save as a varibale 'the_force' in the form of an integer(eg. 25)

```python
the_force = int((star_wars.shape[0] / licensed.shape[0])*100)
print(the_force)
```

```python
# Data cleaning needs to be done, so lets check is there null values in set_num columns
merged[merged['set_num'].isnull()].shape
```

```python
# so lets drop 
licensed = licensed.dropna(subset=['set_num'])
the_force = float((star_wars.shape[0] / licensed.shape[0])*100)
print(the_force)
```

ANSWER : 51.65% 

Question 2: In which year was Star Wars not the most popular licensed theme(in terms of number of sets released that year) ?
save your answer as a variable 'new_era' in the form of an integer (eg.2012)

```python
licensed_sorted = licensed.sort_values('year')
licensed_sorted['count'] = 1
summed_df = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
summed_df.head()
```

```python
max_df = summed_df.sort_values('count', ascending=False).drop_duplicates('year')
max_df.sort_values('year',inplace=True)
max_df.head(100)
```

Answer: new_era = 2017
